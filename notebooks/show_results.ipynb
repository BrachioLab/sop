```python
import matplotlib.pyplot as plt
import json
import os

# EXPLAINER_NAMES = ['lime', 'archipelago', 'rise', 'shap', 'intgrad', 'gradcam', 'sop']
EXPLAINER_NAMES = ['lime', 'rise', 'shap', 'intgrad', 'gradcam', 'sop']

exp_dir = '../exps/imagenet_m_2h/best/results'

metric = 'consistency'

consistencies = []
accuracies = []
for expln_name in EXPLAINER_NAMES:
    with open(os.path.join(exp_dir, f'{expln_name}_{metric}.json'), 'rt') as input_file:
        expln_results = json.load(input_file)
        accuracies.append(expln_results['accuracy'])
        consistencies.append(expln_results['consistency'])
```

```python
plt.bar(EXPLAINER_NAMES, consistencies)
for i, value in enumerate(consistencies):
    plt.text(i, value, str(value), ha='center')
plt.title('Consistency')
```

```python
plt.bar(EXPLAINER_NAMES, accuracies)
for i, value in enumerate(accuracies):
    plt.text(i, value, str(value), ha='center')
plt.title('Accuracy')
```
